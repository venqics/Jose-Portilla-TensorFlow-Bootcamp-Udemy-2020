#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
vertices = np.array([ [250,700], [425,400], [600,725]],np.int32)
```

```python
vertices
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.polylines(img_rgb, [pts], isClosed = True, color = (0,0,255), thickness = 20)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(img_rgb, [pts],color = (0,0,255))
plt.imshow(img_rgb)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import cv2
import numpy as np

def create_circle(event,x,y, flags, param):

    if event == cv2.EVENT_RBUTTON:

        cv2.circle(img, (x,y), 100, (0,0,255), thickness = 10)

img = cv2.imread(r'/content/drive/My Drive/Colab Notebooks/data_science_projects/jose_portilla_tensorflow_Udemy/Computer_vision_with_open_CV_course/DATA/dog_backpack.jpg')

cv2.namedWindow(winname = 'dog')

cv2.setMouseCallback('dog', create_circle)

while True:

    cv2.imshow('dog',img)

    if cv2.waaitkey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
